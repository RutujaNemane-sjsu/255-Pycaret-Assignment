```python
# Loading dataset
from pycaret.datasets import get_data
data = get_data('wholesale')
```

```python
# Initial setup of functional API
from pycaret.clustering import *
s = setup(data, session_id = 123)
```

```python
# Using Object Oriented API
from pycaret.clustering import ClusteringExperiment
experiment = ClusteringExperiment()
```

```python
# setup of Object Oriented API
experiment.setup(data, session_id = 123)
```

```python
# training kmeans model
kmeans = create_model('kmeans')
```

```python
# training meanshift model
kmeans = create_model('meanshift')
```

```python
# Assigning clusters to the training data
kmeans_cluster = assign_model(kmeans)
kmeans_cluster
```

```python
# Analyzing the model
plot_model(kmeans, plot = 'cluster')
```

```python
# evaluating the model
evaluate_model(kmeans)
```

```python
# prediction
pred = predict_model(kmeans, data=data)
```
